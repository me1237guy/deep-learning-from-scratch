```python
import numpy as np
import matplotlib.pylab as plt
```

$$f(x)=\frac{1}{1+exp(-x)}$$

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(start=-5, stop=5.0, step=0.2)
y = sigmoid(x)
```

```python
plt.plot(x, y,'.-')
plt.ylim(-0.1, 1.1)
plt.xlabel('x')
plt.ylabel(r'$f(x)=\frac{1}{1+\exp(-x)}$')
```
